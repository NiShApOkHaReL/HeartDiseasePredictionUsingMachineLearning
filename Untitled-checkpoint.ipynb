```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('cardio', axis=1)
y = df['cardio']
```

```python
from sklearn.model_selection import train_test_split

# Splitting into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Initialize the Decision Tree Classifier
model = DecisionTreeClassifier()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = model.predict(X_test)
```

```python
print('Train Accuracy: ', model.score(X_train,y_train))
print('Test Accuracy: ', model.score(X_test,y_test))
```

# Handling overfitting 

## Stratified Train- Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y, test_size=0.2)
```

```python
model.fit(X_train,y_train)

print('Train Accuracy: ', model.score(X_train,y_train))
print('Test Accuracy: ', model.score(X_test,y_test))
```

## Stratified K-fold Cross Validation

```python
from sklearn.model_selection import StratifiedKFold

X = df.drop('cardio', axis=1).values
y = df['cardio'].values

kfold = StratifiedKFold(n_splits=5, random_state=0, shuffle=True)

for train_index, test_index in kfold.split(X,y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

model.fit(X_train,y_train)

print('Train Accuracy: ', model.score(X_train,y_train))
print('Test Accuracy: ', model.score(X_test,y_test))
```

## Hyperparameter tuning with GridSearchCV

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {'max_depth': [5, 10, 20],'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]}

model = DecisionTreeClassifier()

grid_search = GridSearchCV(model, param_grid, cv=5)

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print('Best Parameters: ', best_params)
```

```python
best_model = DecisionTreeClassifier (**best_params)

best_model.fit(X_train, y_train)

print('Train Accuracy: ', best_model.score(X_train,y_train))
print('Test Accuracy: ', best_model.score(X_test,y_test))
```

```python
# There are only 2 data points above 200 cm so those will be removed as well.
df = df[df["height"] < 201]
```

3. Weight

```python
# Adding a new parameter for Body Mass Index (bmi), a metric used to classify weight per height class.
# This will be a better indicator of health using weight and height.
# BMI = kg/m2

df["bmi"] = df["weight"] *10000 / ((df["height"])**2)
```

```python
# Now the bmi can be split into classes of bmi, which would help understand the data more.

bmi_ranges = [
    (df["bmi"] <= 15),                                 # 0 - Anorexic
    (df["bmi"] > 15) & (df["bmi"] <= 18.5),       # 1 - Underweight
    (df["bmi"] > 18.5) & (df["bmi"] <= 25),       # 2 - Normal weight
    (df["bmi"] > 25) & (df["bmi"] <= 30),         # 3 - Overweight
    (df["bmi"] > 30) & (df["bmi"] <= 35),         # 4 - Obesity Class I
    (df["bmi"] > 35) & (df["bmi"] <= 40),         # 5 - Obesity Class II
    (df["bmi"] > 40)                                   # 6 - Obesity Class III 
]

classes = [0, 1, 2, 3, 4, 5, 6]

df["bmi_class"] = np.select(bmi_ranges, classes)
```

```python
# The bmi class 0 (< 15) alludes to an extremely severe case of anorexia in the patient [3]. This will be created as a new label to account for the condition and to assess if it affects the cardio column.
```
